```python
import yaml

object_names = []
for i in range(1,3):
    filename = 'pick_list_{}.yaml'
    with open('pick_list_1.yaml','r') as stream:
        data = yaml.load(stream)
        for i in range(len(data['object_list'])):
            object_names.append(data['object_list'][i]['name'])

final = set(object_names)
print(final)
```
